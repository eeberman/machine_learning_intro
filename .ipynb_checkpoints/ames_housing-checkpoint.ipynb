```python
null_counts = df.isnull().sum()
null_counts= null_counts[null_counts > 0]
print(null_counts)
```

```python
# Select feature names containing 'sf' or 'area'
square_footage_features = [col for col in df.columns if 'sf' in col.lower() or 'area' in col.lower()]

square_footage_features = [col for col in square_footage_features if col != 'Lot Area']

square_footage_features = [col for col in square_footage_features if col != 'Mas Vnr Area']


# Print the selected feature names
print(square_footage_features)
```

```python
df['square_footage_total'] = df[square_footage_features].sum(axis=1)

print(df[['square_footage_total']].head())
```

```python
df.describe()
```

```python
# Calculate the area lot usage as the ratio of square_footage_total to Lot Area
df['area_lot_usage'] = df['square_footage_total'] / df['Lot Area']

# Print the first few rows to verify the new feature
print(df[['area_lot_usage']].head())
```

```python
df.describe()
```

```python
# Drop rows that have more than a certain number of missing values
threshold = 5  # Adjust the threshold as needed
df_cleaned = df.dropna(thresh=df.shape[1] - threshold)

# Check the shape of the cleaned dataset
print(df_cleaned.shape)
```
