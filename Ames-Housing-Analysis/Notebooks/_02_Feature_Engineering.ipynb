```python
import pandas as pd
# Load the dataset 
raw_data_url = "https://raw.githubusercontent.com/eeberman/machine_learning_intro/master/Ames-Housing-Analysis/Data/AmesHousing.csv"
#df = pd.read_csv(raw_data_url)
df = pd.read_csv(r"C:\Users\elieberm1\Documents\Python Scripts\machine_learning_self\Ames-Housing-Analysis\Data\Raw\AmesHousing.csv")
print(df.head())
```

```python
# Define the square footage features relevant to living space
square_footage_features = [
    'Gr Liv Area',  # Above ground living area
    'Total Bsmt SF',  # Total basement area
    '1st Flr SF',  # First floor area
    '2nd Flr SF',  # Second floor area, if applicable
    'BsmtFin SF 1',  # Finished basement area 1
    'BsmtFin SF 2',  # Finished basement area 2
    'Bsmt Unf SF',  # Unfinished basement area
    'Low Qual Fin SF'  # Low quality finished area
]

# Calculate the total square footage for living space
df['square_footage_total'] = df[square_footage_features].sum(axis=1)

# Print the selected feature names
print("Features included in square_footage_total:")
print(square_footage_features)

# Print the first few rows of the calculated total square footage
print(df[['square_footage_total']].head(10))
# Check and redescribe the dataset with square_footage_total included
df.describe()
```

```python
# Select all numerical features (int and float)
numerical_features = df.select_dtypes(include=[int,float])

# Print selected features
print("Numerical features in the dataset:")
print(numerical_features.columns.tolist())
```

```python
standardize_features = ['Lot Area', 'BsmtFin SF 1','BsmtFin SF 2', 'Bsmt Unf SF', 
'Total Bsmt SF', '1st Flr SF', '2nd Flr SF', 'Low Qual Fin SF', 'Gr Liv Area', 
'Garage Area', 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 
'Screen Porch', 'Pool Area', 'square_footage_total']
# Print selected features
print("Features to standardize in the dataset:")
print(standardize_features)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler = StandardScaler()

# Fit, transform and convert selected features to a DataFrame
X_scaled = scaler.fit_transform(df[standardize_features])
X_scaled_df = pd.DataFrame(X_scaled, columns=standardize_features)

# Verify the scaling 
print(X_scaled_df.describe())
```

```python
# Identify non-standaized features and select them from the dataset
non_standardized_features = [col for col in df.columns if col not in standardize_features]
df_non_standardized = df[non_standardized_features]
print(df_non_standardized.columns.tolist())
```

```python
# Combine the features and verify the combination
df_combined = pd.concat([X_scaled_df, df_non_standardized.reset_index(drop=True)], axis=1)
print(df_combined.head())
```

```python
# Find null datapoints
null_counts = df_combined.isnull().sum()
null_counts = null_counts[null_counts > 0]
null_counts = null_counts.sort_values(ascending=False)
print(null_counts)
```

```python
# Handle missing values for categorical features that are absent according to data description

df_combined = df_combined.fillna({'Garage Qual': 'No Garage', 'Alley':'No Alley', 'Misc Feature':'No Misc Feature', 'Pool QC': 'No Pool',
                        'Fence': 'No Fence', 'Mas Vnr Type': 'No Mas Vnr', 'Fireplace Qu': 'No Fireplace', 'Garage Finish': 'No Garage', 
                         'Garage Cond': 'No Garage', 'Garage Type': 'No Garage', 'Bsmt Exposure' : 'No Basement', 
                         'BsmtFin Type 2': 'No Basement', 'BsmtFin Type 1': 'No Basement', 'Bsmt Qual': 'No Basement',
                        'Bsmt Cond': 'No Basement'})


# Find null datapoints
null_counts = df_combined.isnull().sum()
null_counts = null_counts[null_counts > 0]
null_counts = null_counts.sort_values(ascending=False)
print(null_counts)
```

```python
# Check the remaining null columns to confirm if a 0 can be inputted
columns_of_interest = [
    'Electrical'
]


rows_with_nulls = df_combined[df_combined[columns_of_interest].isnull().any(axis=1)]
pd.set_option('display.max_columns', None)

print(rows_with_nulls)
```

```python
# Input 0s when appropriate 
zero_value_features = ['Lot Frontage', 'Mas Vnr Area', 'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Half Bath', 'Bsmt Full Bath', 'Garage Area', 'Garage Cars', 'Garage Yr Blt']


df_combined[zero_value_features] = df_combined[zero_value_features].fillna(0)
# Find null datapoints
null_counts = df_combined.isnull().sum()
null_counts = null_counts[null_counts > 0]
null_counts = null_counts.sort_values(ascending=False)
print(null_counts)
```

```python
# Impute the missing value in 'Electrical' with the mode 
df_combined['Electrical'] = df_combined['Electrical'].fillna(df_combined['Electrical'].mode()[0])
# Find null datapoints
null_counts = df_combined.isnull().sum()
null_counts = null_counts[null_counts > 0]
null_counts = null_counts.sort_values(ascending=False)
print(null_counts)
```

```python
# Drop the 'PID' column from and verify the column has been removed
df_combined = df_combined.drop(columns=['PID'])


print(df_combined.columns)
```

```python
# Map numerical values to quality scores 
quality_mapping = {
    'No': -1,  
    'Po': 1,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

quality_features = ['Exter Qual', 'Exter Cond', 'Bsmt Qual', 'Bsmt Cond', 
                      'Fireplace Qu', 'Garage Qual', 'Garage Cond', 'Pool QC']

for feature in quality_features:
    df_combined[feature] = df_combined[feature].map(quality_mapping).fillna(df_combined[feature])

# Verify the mapping and confirm there are no null values
print(df_combined[quality_features].head())
null_counts = df_combined.isnull().sum()
null_counts = null_counts[null_counts > 0]
null_counts = null_counts.sort_values(ascending=False)
print(null_counts)
```

```python
# Change the "No ____" values to -1
features_with_no_prefix = ['Garage Qual', 'Alley', 'Misc Feature', 'Pool QC', 'Fence', 
                           'Mas Vnr Type', 'Fireplace Qu', 'Garage Finish', 'Garage Cond', 
                           'Garage Type', 'BsmtFin Type 2', 
                           'BsmtFin Type 1', 'Bsmt Qual', 'Bsmt Cond']


for feature in features_with_no_prefix:
    df_combined[feature] = df_combined[feature].apply(lambda x: -1 if isinstance(x, str) and x.startswith('No') else x)

# Special carve out for no exposure vs no basement
df_combined['Bsmt Exposure'] = df_combined['Bsmt Exposure'].apply(
    lambda x: -1 if x == 'No Basement' else (0 if x == 'No Exposure' else x)
)

print(df_combined[features_with_no_prefix + ['Bsmt Exposure']].head())
```

```python
# Create a binary indicator for the presense of a pool
df_combined['Has_Pool'] = df_combined['Pool Area'].apply(lambda x: 1 if x > 0 else 0)
print(df_combined[['Pool Area', 'Has_Pool']].tail())
```

```python
print(df_combined.describe())
```

```python
df_combined['House_Age'] = df_combined['Yr Sold'] - df_combined['Year Built']
df_combined['Is_Remodeled'] = (df_combined['Year Remod/Add'] != df_combined['Year Built']).astype(int)
df_combined['Total_Bathrooms'] = (df_combined['Full Bath'] + 
                                  0.5 * df_combined['Half Bath'] +
                                  df_combined['Bsmt Full Bath'] +
                                  0.5 * df_combined['Bsmt Half Bath'])
df_combined['Total_Porch_SF'] = (df_combined['Open Porch SF'] +
                                 df_combined['Enclosed Porch'] +
                                 df_combined['3Ssn Porch'] +
                                 df_combined['Screen Porch'])
df_combined['Has_Garage'] = df_combined['Garage Area'].apply(lambda x: 1 if x > 0 else 0)
print(df_combined.describe())
```
